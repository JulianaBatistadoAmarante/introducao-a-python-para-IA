Desafio - Altere os dados para prever:

Horas Estudo	Aprovado
1	0
2	0
3	0
4	1
5	1
6	1
Crie um modelo que responda:

Um aluno que estudou 5 horas será aprovado?

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier

dados = {
    "horas_de_estudos": [1, 2, 3, 4, 5, 6],
    "aprovado": [0, 0, 0, 1, 1, 1]
}

# Cria a tabela
df = pd.DataFrame(dados)

# Entradas
x = df[["horas_de_estudos"]]

# Saídas
y = df["aprovado"]

# Cria o modelo
modelo = DecisionTreeClassifier()

# Treina o modelo
modelo.fit(x, y)

# Nova pessoa
nova_pessoa = [[5]]

# Faz previsão
resultado = modelo.predict(nova_pessoa)

print(resultado)
```
